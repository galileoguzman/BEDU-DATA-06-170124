### Analisis de datos

Saber cuales de los diferentes sitios web (sources) tinen mas interaccion

```python
# Seleccionamos las columnas base para nuestro analisis
COLUMNS = [
    'source_name',
    'author',
    'engagement_reaction_count',
    'engagement_comment_count',
    'engagement_share_count',
    'engagement_comment_plugin_count'
]

# Limpiar todas aquelas filas que tienen un valor nulo en las columnas seleccionadas
data = data[COLUMNS].dropna(axis=0, how='any')
```

```python
# Agrupamos nuestro dataframe por la columna source_name y sumamos cada fila por su valor en la columna engagement_reaction_count
engagement_reaction_count = data.groupby('source_name')['engagement_reaction_count'].sum()
engagement_reaction_count.to_frame()
```

```python
engagement_share_count = data.groupby('source_name')['engagement_share_count'].sum()
engagement_share_count.to_frame()
```

```python
# extraer nueva serie
highest_source = data['source_name'].value_counts()
highest_source.to_frame()
```

### Visualizacion de datos

```python
# Seleccionamos el valor de la primera columna (indice) y lo convertimos a un elemento de lista por cada valor en la serie
highest_source_labels = list(highest_source.index.values)
print(highest_source_labels)
```

```python
# Crear grafica de pastel
plt.pie(
    highest_source,
    labels=highest_source_labels,
    autopct='%1.1f%%'
)
plt.title('Fuentes de informacion')
plt.show()
```
